# AirBnB Rome Listings Analysis


## Data Set Exploration

#### -Reading AirBnb Rome listings data
#### -Casting data columns as a datetime format
#### -Checking missing values, calculating minimum, maximum, and avergae for each numeric field

```python
import pandas as pd

listings = pd.read_csv("listings.csv")

listings
```

```python
listings.info()
```

```python
listings["host_since"] = pd.to_datetime(listings["host_since"])
```

```python
listings.info()
```

```python
rome_listings = listings.loc[:, ["host_since", "neighbourhood_cleansed", "accommodates", "price"]]
```

```python
rome_listings.info()
```

```python
rome_listings.isna().sum()
```

```python
rome_listings.describe()
```

## Prepare for Visualization

#### -Creating a DataFrame rome_listings_neighborhood. Grouping the Rome listings by neighborhood and calculating the average price for each. Sorting by price in ascending order.
#### -Creating a DataFrame rome_listings_accommodates. Filtering down to the most expensive neighborhoods in Rome. Grouping it by accommodates and calculating the average price fro each. Sort by price in ascending order.
#### -Creating a DataFrame rome_listings_over_time. Grouping the data by the year component of host_since. Calculating a count of rows to get the number of new hosts for each year, and the average price of listings for each year.

```python
rome_listings_neighborhood = ( 
    rome_listings
    .groupby("neighbourhood_cleansed")
    .agg({"price":"mean"})
    .sort_values("price")
)

rome_listings_neighborhood.tail()
```

```python
rome_listings_accommodates = (
    rome_listings
    .query("neighbourhood_cleansed == 'I Centro Storico'")
    .groupby("accommodates")
    .agg({"price":"mean"})
    .sort_values("price")
)
rome_listings_accommodates.head()
```

```python
rome_listings_over_time = (
    rome_listings
    .set_index("host_since")
    .resample("Y")
    .agg({
        "neighbourhood_cleansed" : "count",
        "price" : "mean"
    })
)

rome_listings_over_time.head()
```

## Visualize the data

#### -Building a horizontal bar chart of average rent price by the neighborhood.
#### -Building a horizontal bar chart of average price by accommodates in the most expensive neighborhood.
#### -Building line charts of new hosts per year and average price by year.
#### -Plotting both time series in dual axis line chart.

```python
import seaborn as sns

(rome_listings_neighborhood
 .plot
 .barh(
     title="Average Listing Price by Rome Neighborhood", 
     xlabel="Price Per Night (USD)", 
     ylabel="Neighborhood",
     legend=None
 )
)

sns.despine()
```

```python
(rome_listings_accommodates
 .plot
 .barh(
     title="Average Listing Price by Accommodation Number", 
     xlabel="Price Per Night (USD)", 
     ylabel="Accommodation Capacity",
     legend=None
 )
)

sns.despine()
```

```python
rome_listings_over_time["neighbourhood_cleansed"].plot(
    ylabel="New Hosts",
    title="New AirBnB Hosts in Rome Over Time"
)

sns.despine()
```

```python
rome_listings_over_time["price"].plot(
    ylabel="Average Price (USD)",
    title="Average AirBnB Price in Rome Over Time"
)

sns.despine()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(
    rome_listings_over_time.index,
    rome_listings_over_time["neighbourhood_cleansed"],
    label="New Hosts",
    c="purple"
)

ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax2.plot(
    rome_listings_over_time.index,
    rome_listings_over_time["price"],
    label="Average Price",
)
ax2.set_ylim(0)

ax2.set_ylabel("Average Price (USD)")

ax.set_title("Increase of  New Hosts Number and Average Accommodation Price after COVID")

ax.legend()
ax2.legend()
```
